Bài báo này giống như một lời tuyên chiến mà phái Bayes gửi đến cho phái frequentist, nó báo hiệu một thời đại mới khi cách suy diễn thống kê truyền thống dựa vào giá trị p càng lúc càng bị nghi ngờ và hậu quả tất yếu là giá trị p phải bị đào thải. ### Chuẩn bị priors và likelihood cho mô hình Bayes

Phương pháp Bayes t test của Kruschke khởi đầu bằng ý tưởng: dùng một phân phối Student's t để mô tả biến ngẫu nhiên (thí dụ: DLCO, outcome trong mô hình). Điều này mới nghe có vẻ kì lạ, vì ta vẫn quen với giả định biến kết quả có phân bố Gaussian (chuẩn) trong t-test cổ điển, tuy nhiên đây lại là một giải pháp tiện lợi vì phân phối t thực chất cũng khá tương đồng với phân phối Gaussian, thí dụ tính đối xứng, mặt khác nó linh động hơn so với Gaussian vì có đến 3 tham số bao gồm 1 tham số kiểu hình (hay còn gọi là độ tự do) là $\nu$ (nu):

Biểu đồ sau đây trình bày sự biến hình của phân phối t tùy theo nu:

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/4/41/Student_t_pdf.svg/325px-Student_t_pdf.svg.png">

Giá trị $\nu$ càng cao thì phân phối t càng xấp xỉ gần với phân phối chuẩn, $\nu$ càng thấp thì phần đuôi của đồ thị càng dày.

Phân phối t ít bị ảnh hưởng bởi outlier so với Gaussian, nhưng nó vẫn cho phép xác định vị trí trung tâm (mean) bằng tham số $\mu$ (mu) và phương sai với tham số $\lambda$ (lambda)

$$f(x|\mu,\lambda,\nu) = \frac{\Gamma(\frac{\nu + 1}{2})}{\Gamma(\frac{\nu}{2})} \left(\frac{\lambda}{\pi\nu}\right)^{\frac{1}{2}} \left[1+\frac{\lambda(x-\mu)^2}{\nu}\right]^{-\frac{\nu+1}{2}}$$

Ghi chú: phân phối t của pyMC3 sử dụng tham số $\lambda$ = $\sigma^{(-2)}$  với $\sigma$ là độ lệch chuẩn

Từ giả định này, ta có thể ước tính giá trị DLCO cho từng phân nhóm bằng 2 hàm likelihood :
          
$$DLCO^{(fibrose)}_i \sim T(\nu, \mu_1, \sigma_1)$$

$$DLCO^{(control)}_i \sim T(\nu, \mu_2, \sigma_2)$$

Mục tiêu của chúng ta là xác định phân phối hậu nghiệm cho 3 tham số: $\mu$, $\sigma$ và $\nu$. 
Một khi đã làm được điều này, tìm ra phân phối hậu nghiệm của khác biệt giữa $\mu_1$ và $\mu_2$ là điều hoàn toàn khả thi, và suy diễn thống kê sẽ được thực hiện sau đó với giả thuyết vô hiệu là khác biệt này lớn hoặc nhỏ hơn 0

Ta cần phải có giả thuyết tiền định (prior) cho mỗi tham số. Để đơn giản, ta giả định tham số $\nu$ là như nhau cho 2 phân nhóm, và có prior là phân phối Exponential, 

còn $\mu_1$ và $\mu_2$ là số thực nên có prior là một phân phối Gaussian, 

$$\mu_k \sim Gaussian(\bar{x}, 2s)$$

cuối cùng ta dùng 1 prior rỗng (uniform) cho tham số sigma:

$$\mu_k \sim Uniform(1,10)$$

###  Khai báo prior cho tham số µ, từ đó ước tính trung bình cho 2 phân nhóm

```python
μ_m = df.DLCO.mean()
μ_s = df.DLCO.std() * 2

with pm.Model() as model:
    F_mean = pm.Normal('Fibrosis_mean', μ_m, sd=μ_s)
    N_mean = pm.Normal('Control_mean', μ_m, sd=μ_s)
    
model
```

```python
pm.kdeplot(np.random.normal(μ_m,μ_s,size=10000),shade=0.5);
```

### Khai báo prior cho tham số σ, từ đó ước tính độ lệch chuẩn của 2 phân nhóm

```python
σ_low = 1
σ_high = 10

with model:
    F_std = pm.Uniform('Fibrosis_std', lower=σ_low, upper=σ_high)
    N_std = pm.Uniform('Control_std', lower=σ_low, upper=σ_high)

model
```

```python
pm.kdeplot(np.random.uniform(σ_low,σ_high,size=10000),shade=0.5);
```

###  Khai báo prior cho tham số $\nu$

```python
with model:
    ν = pm.Exponential('ν_minus_one', 1/29.) + 1
```

```python
pm.kdeplot(np.random.exponential(30, size=1000), shade=0.5);
```

#### Từ độ lệch chuẩn, tạo ra 2 tham số  $\lambda$ , sau đó ước tính giá trị DLCO cho mỗi phân nhóm bằng hàm likelihood sử dụng phân phối Student t, khai báo dữ liệu quan sát là 2 arrays:

```python
with model:
    λ1 = F_std**-2
    λ2 = N_std**-2
    
    Fibrosis = pm.StudentT('Fibrosis', nu=ν, mu=F_mean, lam=λ1, observed=dlco_fibrosis)
    Control = pm.StudentT('Control', nu=ν, mu=N_mean, lam=λ2, observed=dlco_control)
```

### Sản phẩm cuối cùng: khác biệt giữa 2 trung bình, 2 độ lệch chuẩn và kích thước hiểu ứng (effect-size)

```python
with model:
    diff_of_means = pm.Deterministic('difference of means', N_mean - F_mean)
    diff_of_stds = pm.Deterministic('difference of stds', N_std - F_std)
    effect_size = pm.Deterministic('effect size', 
                                   diff_of_means / np.sqrt((F_std**2 + N_std**2) / 2))
```

Tóm tắt toàn bộ mô hình Bayes:

```python
model
```

### Bắt đầu lấy mẫu MCMC

Sau khi đóng gói xong mô hình Bayes, ta suy diễn Bayes bằng cách tạo ra 4 chuỗi MCMC, mỗi chuỗi gồm 2500 lượt

```python
with model:
    trace = pm.sample(draws=2000,n_init=1000,chains=4,tuning=1000,cores=1)
```

Trước hết ta khảo sát phân phối hậu nghiệm của các tham số trong mô hình :

Mô hình Bayes cung cấp cho ta lần lượt:

Fibrosis mean = trung bình DLCO của nhóm Xơ phổi
Control_mean = trung bình DLCO của nhóm chứng,

tương tự, độ lệch chuẩn DLCO của 2 phân nhóm: Fibrosis_sd và Control_sd

Tham số Nu cho phép hình dung về phân bố của DLCO gần với phân bố chuẩn đến mức nào.

Tuy nhiên, kết quả ta cần quan tâm chính là khác biệt giữa 2 phân nhóm: trung bình và độ lệch chuẩn, cũng như effect_size.

```python
# Hàm này dùng để tính Median cho chuỗi MCMC
def trace_median(x):
    return pd.Series(np.median(x),name='Median')

# Hàm này dùng để vẽ biểu đồ phân phối hậu nghiệm và chuỗi MCMC cho mô hình GLM
def plot_traces(traces):
    ax = pm.traceplot(traces,figsize=(12,len(traces.varnames)*1.5),
        lines={k: v['Median'] for k, v in pm.summary(traces, stat_funcs=[trace_median]).iterrows()})

    for i, mn in enumerate(pm.summary(traces, stat_funcs=[trace_median])['Median']):
        ax[i,0].annotate('{:.2f}'.format(mn), xy=(mn,0), xycoords='data'
                    ,xytext=(5,10), textcoords='offset points', rotation=90
                    ,va='bottom', fontsize='large', color='#AA0022')
        
plot_traces(trace)
```

Theo kết quả này, ta có phân phối hậu nghiệm của khác biệt trung bình giữa 2 phân nhóm là 1 phân phối Gaussian, với vị trí trung tâm nằm ở giá trị 6.911 đơn vị DLCO, và mật độ cao nhất (phân vị 2.5 - 97.5) từ -0.148 đến 14.339 đơn vị; 

Effect_size của hiệu ứng "phân nhóm" đối với DLCO: trung bình = 0.845 , mật độ cao nhất trong khoảng -0.038 đến 1.748

```python
pm.summary(trace,alpha=0.05).round(3)
```

```python
pm.forestplot(trace, varnames=['Fibrosis_mean',
                               'Control_mean',
                               'Fibrosis_std',
                              'Control_std',
                              'difference of means',
                              'difference of stds']);
```

Thư viện pyMC3 chịu ảnh hưởng rất lớn từ phong cách của J.Kruschke, với biểu đồ phân phối hậu nghiệm như sau: 

```python
pm.plot_posterior(trace, varnames=['Fibrosis_mean','Control_mean', 'Fibrosis_std', 'Control_std', 'ν_minus_one'],
                  color='gold');
```

Trong biểu đồ này, có thể thấy 97.1 % phân phối hậu nghiệm của khác biệt trung bình nằm trên ngưỡng 0

```python
pm.plot_posterior(trace, varnames=['difference of means','difference of stds', 'effect size'],
                  ref_val=0,
                  color='gold');
```
