Trong bài này, Nhi sẽ áp dụng cùng phương pháp Suy diễn Bayes, nhưng bằng một họ phân phối khác đó là Bernoulli để trả lời 2 câu hỏi: 

1) Xác suất thành công của CNGD khi áp dụng cho một học sinh bất kì là bao nhiêu ?
2) Có sự khác biệt ý nghĩa về xác suất thành công này giữa CNGD và chương trình bình thường hay không ?

Phân phối Bernoulli cũng là một quy luật phân phối cho biến rời rạc như Binomial, nó cũng khởi phát từ một hành động thử nghiệm mà kết quả chỉ có thể nhận 1 trong 2 giá trị (thí dụ 0 hoặc 1). Nhưng khác với Binomial ước lượng số lần thu được 1 kết qủa xác định (thí dụ =1) trong n lần thử, kết quả của phân phối Bernoulli ước lượng chính giá trị 0/1 cho 1 lần thử duy nhất (do đó có thể xem Bernoulli là trường hợp đặc biệt của Binomial với n=1). 

Hàm mật độ xác suất cho phân phối Bernoulli rất đơn giản

$$ f(k;p) = \left\{\begin{matrix} p  (k=1)
\\ 
1-p (k=0)
\end{matrix}\right. $$

Phân phối Bernoulli chỉ có 1 tham số là p, chỉ xác suất thành công (x nhận giá trị 1). 

Trung bình của phân phối Bernoulli cũng chính là p.

Khi áp dụng phân phối Bernoulli, ta đang muốn suy diễn thống kê về xác suất thành công (p) của CNGD cho 1 cá thể học sinh bất kì; và ta cũng có thể so sánh p giữa CNGD và chương trình hiện hành. Nhưng các bạn đặc biệt lưu ý điều này, p của Bernoulli không còn đồng nghĩa với "tỉ lệ thành công" như trong phân bố Binomial (vốn có trung bình = n.p). Do đó, từ thông tin mà báo Lao Động cung cấp, chúng ta phải đi qua một bước mô phỏng để tạo ra dữ liệu có n cá thể dựa vào phân phối Bernoulli, với giả định p = tỉ lệ đạt chuẩn trên n. Sau khi mô phỏng, chúng ta sẽ thấy là trung bình của dữ liệu mô phỏng sẽ rất gần, nhưng tương đương với tỉ lệ giả định ban đầu. Điều này cũng dễ hiểu và hợp lý, vì trên thực tế ta không bao giờ có thể xác định được chắc chắn xác suất p này cả, nó là yếu tố bất định (stochastic), ta chỉ có thể ước lượng nó trong 1 mẫu khảo sát một cách ngẫu nhiên.

## Minh họa quy trình mô phỏng để tạo dữ liệu từ phân phối Binomial

Ta giả định p_reading là tỉ lệ đạt chuẩn môn đọc trên n_obs = 478 học sinh được dạy bằng CNGD, và kết quả đạt chuẩn hay không của từng học sinh tuân theo phân phối Bernoulli, ta có thể mô phỏng 478 kết quả ngẫu nhiên với 1= đạt chuẩn, 0= không đạt:

```python
# Giả định tham số p tương đương với tỉ lệ thành công (p_reading) trên 478 trường hợp

p_reading = 0.9603  # Thực ra chúng ta không biết giá trị của p

# Mô phỏng giá trị quan sát với giả định kết quả thử nghiệm tuân theo quy luật Bernoulli với p=0.9603
n_obs = 478

reading_obs = stats.bernoulli.rvs(p_reading, size=n_obs)

print(reading_obs[:200])
```

Tỉ lệ thành công chính là trung bình của 478 giá trị này, và nó cũng chính là p; như ta thấy, giá trị này không giống như kết quả trên bài báo Lao Động. Tham số p mà ta muốn ước tính không đơn giản chỉ là tỉ lệ trong 1 bảng chéo.

```python
proportion=100*np.sum(reading_obs)/n_obs

proportion
```

```python
np.mean(reading_obs)
```

```python
pm.kdeplot(reading_obs,color="blue",shade=0.3)
plt.ylabel('Mật độ phân bố', fontsize=14)
plt.xlabel('Xác suất đạt chuẩn môn đọc ở mỗi học sinh', fontsize=14)
plt.vlines(p_reading, 0, 17.5, linestyle="--", label="Tỉ lệ thực tế: 0.9603",color="green")
plt.vlines(np.mean(reading_obs), 0, 17.5, linestyle="--", label="Tỉ lệ mô phỏng: 0.9519",color="red")
plt.title("Phân phối Bernoulli của xác suất đạt chuẩn môn đọc")
plt.legend();
```

```python
print("Tỉ lệ học sinh đạt chuẩn môn Đọc ? %.4f" % proportion)
print("Tỉ lệ này có tương đương với giá trị p hay không ? %s" % (proportion == p_reading))
```

Muốn ước lượng giá trị tham số p, ta phải dùng suy diễn Bayes, với prior Beta(0.5,0.5) và dữ liệu mô phỏng (dĩ nhiên, ta không bao giờ có dữ liệu thực tế trong tay). Mô hình Bayes này cho ra kết quả như sau:

```python
with pm.Model() as model:
    p = pm.Beta('Xác suất p', alpha=0.5, beta=0.5)
    obs = pm.Bernoulli("Quan sát", p, observed=reading_obs)
    trace = pm.sample(draws=3000, n_init=1000,cores=1)
    burned_trace = trace[1000:]
```

```python
pm.kdeplot(np.random.beta(0.5,0.5,size=10000),shade=0.5);
```

```python
def trace_median(x):
    return pd.Series(np.median(x),name='Median')

# Hàm này dùng để vẽ biểu đồ phân phối hậu nghiệm và chuỗi MCMC cho mô hình GLM
def plot_traces(traces):
    ax = pm.traceplot(traces,figsize=(12,len(traces.varnames)*1.5),
        lines={k: v['Median'] for k, v in pm.summary(traces, stat_funcs=[trace_median]).iterrows()})

    for i, mn in enumerate(pm.summary(traces, stat_funcs=[trace_median])['Median']):
        ax[i,0].annotate('{:.3f}'.format(mn), xy=(mn,0), xycoords='data'
                    ,xytext=(5,10), textcoords='offset points', rotation=90
                    ,va='bottom', fontsize='large', color='#AA0022')
        
plot_traces(burned_trace)
```

Theo kết quả này, phân phối hậu nghiệm của p có mật độ cao nhất (97.5%) trong khoảng 0.932 đến 0.97; và trung vị = trung bình = 0.951.

```python
pm.summary(burned_trace, alpha=.05).round(3)
```

Như ta thấy, phân phối hậu nghiệm của p chứa giá trị quan sát được trong nghiên cứu của Trung tâm CNGD = 0.96; nhưng đây chỉ là một trường hợp ngẫu nhiên, trên thực tế ta không bao giờ xác định được p mà chỉ có thể ước lượng. Nếu áp dụng CNGD cho 1 mẫu N học sinh khác, ta có thể sẽ thu được một tỉ lệ thành công khác, dao động từ 0.93 đến 0.97 và giá trị kì vọng khoảng 0.95.

```python
plt.title("Phân phối hậu nghiệm của xác suất đạt chuẩn môn Đọc")
plt.vlines(p_reading, 0, 50, linestyle="--", label="Giá trị quan sát của p")
plt.vlines(np.median(burned_trace["Xác suất p"]), 0, 50, linestyle="-", label="trung vị của pp hậu nghiệm",color="red")
plt.hist(burned_trace["Xác suất p"], bins=100, histtype="stepfilled", normed=True, color="red",alpha=0.3)
plt.xlabel('Xác suất đạt chuẩn môn đọc ở mỗi học sinh', fontsize=14)
plt.ylabel('Mật độ phân phối', fontsize=14)
plt.legend();
```

Trong một suy diễn Bayes khác, giả sử ta quy định một ngưỡng vô hiệu cho p, thí dụ 0.93; ta có thể nói: 97.3 % phân phối hậu nghiệm của p nằm bên phải ngưỡng vô hiệu.

```python
pm.plot_posterior(burned_trace["Xác suất p"], 
                  color='red',
                  alpha=0.3,
                  ref_val=0.93)
plt.title("Phân phối hậu nghiệm p cho môn đọc")
plt.xlabel('Xác suất đạt chuẩn môn đọc ở mỗi học sinh', fontsize=14)
plt.ylabel('Mật độ phân phối', fontsize=14)
```

## Môn Đọc

Ta làm tương tự như trên cho 2 phân nhóm: CNGD và BT (bình thường); ta mô phỏng 2 mẫu dữ liệu :

```python
p_cngd = 0.9603
p_bt = 0.9273

N_cngd = 478
N_bt = 220

obs_cngd = stats.bernoulli.rvs(p_cngd, size=N_cngd)
obs_bt = stats.bernoulli.rvs(p_bt, size=N_bt)
print("Mẫu mô phỏng cho nhóm CNGD: ", obs_cngd[:30], "...")
print("Mẫu mô phỏng cho nhóm BT: ", obs_bt[:30], "...")
```

Tỉ lệ thành công ở từng phân nhóm lần lượt là:

```python
print(np.mean(obs_cngd))
print(np.mean(obs_bt))
```

Ta dựng mô hình Bayes tương tự như trên, với 2 tham số p_cndg và p_bt; cả 2 đều có prior là Beta(0.5,0.5).

Khác biệt dif giữa 2 phương pháp CNGD và BT được tính như một thành phần Xác định (deterministic) trong mô hình (nghĩa là khi có p_cngd, p_bt thì ta sẽ tính được dif)

Suy diễn Bayes bắt đầu...

```python
with pm.Model() as model_Doc:
    
    p_cngd = pm.Beta('xác suất_cngd', alpha=0.5, beta=0.5)
    p_bt = pm.Beta('xác suất_bt', alpha=0.5, beta=0.5)
    
    dif = pm.Deterministic("Khác biệt", p_cngd - p_bt)

    obs_A = pm.Bernoulli("Nhóm CNGD", p_cngd, observed=obs_cngd)
    obs_B = pm.Bernoulli("Nhóm BT", p_bt, observed=obs_bt)

    trace_Doc = pm.sample(draws=5000, n_init=1000,cores=1)
    burned_Doc = trace_Doc[1000:]
```

Kết quả cho thấy: khác biệt trung bình về xác suất thành công trong môn Đọc giữa 2 phương pháp dạy học rất thấp, chỉ có 0.029 (2.9 %) (khoảng xác tín từ -0.008 đến +0.067).

```python
pm.summary(burned_Doc, alpha=.05).round(3)
```

Ở đây ta không có định kiến nào cả với trung tâm CNGD, tuy nhiên nếu ta đặt 1 ngưỡng giá trị vô hiệu cho khác biệt để nghiệm thu chương trình của họ, và ngưỡng này là +0.05, thì ta sẽ thấy chỉ có 14% mật độ phân phối hậu nghiệm cao hơn ngưỡng này.

```python
%matplotlib inline
pm.plot_posterior(burned_Doc['Khác biệt'], color='gold',alpha=0.8,ref_val=0.05)
plt.title("Phân phối hậu nghiệm : Khác biệt xác suất thành công giữa 2 phương pháp")
plt.xlabel('Khác biệt xác suất thành công (Đọc)', fontsize=14)
plt.ylabel('Mật độ phân phối', fontsize=14)
```

```python
p_cngd_samples = burned_Doc["xác suất_cngd"]
p_bt_samples = burned_Doc["xác suất_bt"]
dif_samples = burned_Doc["Khác biệt"]

p_CN = 0.9603
p_BT = 0.9273
```

```python
%matplotlib inline

figsize(12.5, 10)
ax = plt.subplot(311)

plt.xlim(0.8,1)
plt.hist(p_cngd_samples, histtype='stepfilled', bins=50, alpha=0.5,
         label="PP hậu nghiệm của p_cngd", color="red", normed=True)
plt.vlines(p_CN, 0,50,linestyle="--", label="giá trị thực tế p_cngd (cần xác minh)")
plt.legend(loc="upper left")
plt.title("PP hậu nghiệm của xác suất thành công và khác biệt giữa 2 nhóm (Môn Đọc)")
plt.ylabel('Tần suất', fontsize=14)

ax = plt.subplot(312)

plt.xlim(0.8,1)
plt.hist(p_bt_samples, histtype='stepfilled', bins=50, alpha=0.5,
         label="PP hậu nghiệm của p_bt", color="blue", normed=True)
plt.vlines(p_BT, 0,30,linestyle="--", label="giá trị thực tế p_bt (cần xác minh)")
plt.legend(loc="upper left")
plt.ylabel('Tần suất', fontsize=14)

ax = plt.subplot(313)
plt.hist(dif_samples, histtype='stepfilled', bins=50, alpha=0.8,
         label="PP hậu nghiệm của khác biệt", color="gold", normed=True)
plt.vlines(p_CN - p_BT, 0, 25, linestyle="--",
           label="khác biệt thực tế (cần xác minh)")
plt.vlines(0.05, 0, 25, color="red", label="Ngưỡng vô hiệu : +5%")
plt.ylabel('Tần suất', fontsize=14)

plt.legend(loc="upper left");
```

Cả hai phân phối Binomial và Bernoulli đều có thể dẫn chúng ta đến một hướng đi khác là hồi quy logistic, với link function là một hàm sigmoid (logit).
