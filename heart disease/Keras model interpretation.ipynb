Ta viết 1 vòng lặp để chuẩn hóa các biến định lượng liên tục, nhưng giữ nguyên biến định tính (trong bài scikitlearn 1 : Hồi quy, Nhi dùng một pipeline để giải quyết cùng mục tiêu này), đây là 1 cách làm khác.

Tiếp theo, ta dùng hàm train_test_split của scikit-learn để cắt dữ liệu X,y và X_disp y_disp thành 2 phần: Train subsets chứa 80% (n=528), validation subset = 20% còn lại. Chú ý là tập X_disp được chia theo cùng một quy luật so với tập X.

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=123)

X_disp_train, X_disp_test, y_disp_train, y_disp_test = train_test_split(X_disp,y_disp, test_size=0.2, random_state=123)
```

Một bước nhỏ sau cùng để chuyển giá trị output y thành biến nhị phân (0,1):

```python
encoded_y_train[0:5], encoded_y_valid[0:5]
```

Để có thể diễn giải được bằng SHAP, mô hình Keras được dựng theo một cấu trúc phức tạp hơn so với bình thường. Cụ thể, ở đây ta không code mô hình theo kiểu Sequential, mà sử dụng Function API, trong đó đầu vào, đầu ra và các lớp ẩn được thiết kế riêng biệt theo từng bộ phận trước khi ráp nối với nhau qua hàm Model( ).

Đầu vào của mạng neuron này có nhiều lớp, nó có khả năng phân biệt datatype của dữ liệu đầu vào và xử lý từng features theo luồng riêng biệt, các biến định tính mã hóa số nguyên được xử lý qua các lớp Embedding và flatten, trong khi các biến liên tục được tiếp nhận trực tiếp. Có đến 20 lớp với chỉ 1 neuron để nhận 10 features, để cuối cùng kết nối lại qua 1 lớp concatenate để tạo ra tensor input sau cùng.

Tensor này sau đó được đi qua tiếp phần lõi của mô hình, gồm 2 lớp Dense (128 và 64 neurons, hàm RELU) xen kẽ với 2 lớp dropout tỉ lệ loại bỏ ngẫu nhiên là 0.2, cuối cùng kết quả được xuất ra 1 neuron kích hoạt bằng hàm sigmoid (như một mô hình logistic). 

```python
# Dựng một mô hình keras kiểu "function"

input_els = []
encoded_els = []

# Đầu vào = 5 Lớp Embedding + flatten
for k,dtype in dtypes:
    input_els.append(Input(shape=(1,))) 
    if dtype == "int8":
        e = Flatten()(Embedding(X_train[k].max()+1, 1)(input_els[-1])) # embedding cho feature định tính (int8)
    else:
        e = input_els[-1] # riêng cho feature định lượng (float32)
    encoded_els.append(e)

encoded_els = concatenate(encoded_els)  # Nhập features định tính và định lượng vào cùng tensor

hidden1 = Dense(128, kernel_initializer='normal', activation="relu")(encoded_els) # Lớp ẩn thứ 1: 128 neuron
droplayer1 = Dropout(0.2)(hidden1) # Lớp dropout thứ 1 
hidden2 = Dense(64, kernel_initializer='normal', activation="relu")(droplayer1) # lớp ẩn thứ 2: 64 neuron
droplayer2 = Dropout(0.2)(hidden2) # Lớp dropout thứ 2
outputlayer = Dense(1, kernel_initializer='normal', activation='sigmoid')(droplayer2) # Lớp sau cùng, xuất kết quả

classifier = Model(inputs=input_els, outputs=[outputlayer])

classifier.compile(optimizer="adam", loss='binary_crossentropy', metrics=['accuracy'])

classifier.summary()
```

Mô hình được huấn luyện qua 50 epoches và nhanh chóng đạt accuracy trên tập kiểm định đến 0.8, kết quả này xem như tương đương với một mô hình Random Forest mà Nhi từng làm trong caret.

```python
classifier.fit(
    [X_train[k].values for k,t in dtypes],
    encoded_y_train,
    epochs=50,
    batch_size=128,
    shuffle=True,
    validation_data=([X_valid[k].values for k,t in dtypes], encoded_y_valid) 
)
```

Bây giờ chúng ta qua phần chính của bài thực hành: Diễn giải mô hình phân loại Deep learning nói trên.

Trước hết, Nhi chuẩn bị 1 hàm pred_func có chức năng đưa từng features riêng lẻ vào mạng neuron sâu vừa tạo ở trên. Cấu trúc đặc biệt của lớp Input của mạng neuron này cho phép tiếp nhận riêng biệt mỗi features theo 1 luồng riêng, do đó mô hình của chúng ta có thể hoạt động mà không cần phải có đủ cả 10 features đồng thời, chỉ cần 1 feature mô hình vẫn làm việc được, và nhờ đó ta mới đánh giá được vai trò riêng biệt của từng features.

```python
def pred_func(X):
    return classifier.predict([X[:,i] for i in range(X.shape[1])]).flatten()
```

Hàm KernelExplainer trên toàn bộ tập X_valid sẽ tạo ra phần dữ liệu cơ bản, gọi là explainer để giải thích cơ chế hoạt động của mô hình trên tập validation (test) này

```python
explainer = shap.KernelExplainer(pred_func, X_valid.iloc[:len(X_valid),:])
```

Chú ý: Mỗi hàng trong tập X_valid và X_disp_test là như nhau, cùng một bệnh nhân, tuy nhiên giá trị của features khác nhau: tập X_valid đã bị can thiệp chuẩn hóa và mã hóa, trong khi tập X_disp vẫn giữ nguyên thang đo và đơn vị gốc. Kết quả diễn giải cho case số 30 được trình bày bằng một biểu đồ trực quan, dễ hiểu như sau: Màu xanh gợi ý tất cả các features đều đóng góp kéo lùi xác suất chẩn đoán Có bệnh về giá trị 0.03 theo mức độ mạnh, yếu khác nhau (chiều dài các mũi tên); các giá trị features được liệt kê bên dưới. Như vậy kết quả chẩn đoán của mô hình là chính xác, phù hợp với quan sát thực tế là không có bệnh

```python
# Case N°30 trong testset

print("Thực tế =", y_disp_test[30])
print("Giải thích mô hình : ")
shap_values = explainer.shap_values(X_valid.iloc[30,:], nsamples=500)
shap.force_plot(explainer.expected_value, shap_values, X_disp_test.iloc[30,:])
```

```python
# Case N°50 trong testset

print("Thực tế =", y_disp_test[50])
print("Giải thích mô hình : ")
shap_values = explainer.shap_values(X_valid.iloc[50,:], nsamples=500)
shap.force_plot(explainer.expected_value, shap_values, X_disp_test.iloc[50,:])
```

Ta có thể trình bày kết quả diễn giải bằng biểu đồ waterfall cho từng case hay hàng loạt case :

```python
import waterfall_chart

for i in range(10):
    shap_values = explainer.shap_values(X_valid.iloc[i,:], nsamples=500)
    waterfall_chart.plot(X_disp.columns.values,shap_values, 
                             rotation_value=90, 
                             threshold=0.3,
                             formatting='{:,.3f}',
                             green_color='#f21d6b', red_color='#22a3e8', blue_color='#f7c72a',
                             Title=i+1);
```

Không chỉ diễn giải được cho cá thể, SHAP còn cho phép dựng nhiều biểu đồ rất thú vị cho phép diễn giải được vai trò của các features ở cấp độ quần thể như sau:

```python
shap_values_test = explainer.shap_values(X_valid, nsamples=500)
```

Biểu đồ sau đây trình bày vai trò của từng features cũng như tập hợp của chúng lên xác suất chẩn đoán có bệnh (trục tung) cho 130 cases (trục hoành), chú ý điều thú vị là những cases có chẩn đoán tương tự và phân bố features tương tự được nhóm lại thành nhóm gần nhau.

```python
shap.force_plot(base_value = explainer.expected_value, 
                 shap_values = shap_values_test, 
                 features = X_disp_test,
                 figsize=(20,7))
```

Mức độ quan trọng của mỗi features, đo bằng SHAP values (hiệu ứng lên output) có thể khảo sát bằng 3 loại biểu đồ khác nhau: dotplot, violin plot và barplot
